# Updata using Python

```python
%python

from delta.tables import DeltaTable

# Reference the table by its Unity Catalog name
deltaTable = DeltaTable.forName(spark, "adb_alejandro_research_dev_aue_ws.f1_demo.results_managed")

# Declare the predicate by using a SQL-formatted string.
deltaTable.update(
  condition = "position <=10",
  set = { "points": "21-position" } )
```

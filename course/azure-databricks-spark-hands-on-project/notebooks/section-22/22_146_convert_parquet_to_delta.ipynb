```python
%sql
DESCRIBE SCHEMA EXTENDED f1_demo;
```

```python
%sql
-- Creating a parquet table

CREATE TABLE IF NOT EXISTS adb_alejandro_research_dev_aue_ws.f1_demo.drivers_convert_to_delta (

 driverId INT
,dob DATE
,forename STRING
,surname STRING
,createdDate DATE
,updatedDate DATE

)

USING PARQUET
LOCATION 'abfss://demo@alejandroauedevdl.dfs.core.windows.net/drivers_convert_to_delta'
```


# From parquet file* to delta table

```python
df = spark.table("f1_demo.drivers_convert_to_delta")
```

```python
df.write.format("parquet").mode("overwrite").save("abfss://demo@alejandroauedevdl.dfs.core.windows.net/drivers_convert_to_delta_new")
```

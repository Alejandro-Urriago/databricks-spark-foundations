## Go back to previous version if my current version was corrupted

```python
%sql

-- restablish vacuum retention

SET spark.databricks.delta.retentionDurationCheck.enabled = true;
```

```python
%sql
DESC HISTORY f1_demo.drivers_merge
```

```python
%sql

SELECT * FROM f1_demo.drivers_merge
VERSION AS OF 4;
```

```python
%sql

DELETE FROM f1_demo.drivers_merge WHERE driverId =1;
```

```python
%sql

MERGE INTO f1_demo.drivers_merge AS tgt
USING f1_demo.drivers_merge VERSION AS OF 4 AS src
  ON (tgt.driverId = src.driverId)
WHEN NOT MATCHED THEN 
  INSERT *
```

### Create Catalog and Schemas required for the project

1. Catalog - formula1_dev (without managed location)
1. Schemas - bronze, silver and gold ( with managed location)

```sql
CREATE CATALOG IF NOT EXISTS f1_dev 
MANAGED LOCATION 'abfss://unity-catalog-prod@alejandroauedevdl.dfs.core.windows.net/f1_dev';
```

```sql
USE CATALOG f1_dev;
```

```sql
CREATE SCHEMA IF NOT EXISTS bronze
MANAGED LOCATION 'abfss://bronze@alejandroauedevdl.dfs.core.windows.net/'
```

```sql
CREATE SCHEMA IF NOT EXISTS silver
MANAGED LOCATION 'abfss://silver@alejandroauedevdl.dfs.core.windows.net/'
```

```sql
CREATE SCHEMA IF NOT EXISTS gold
MANAGED LOCATION 'abfss://gold@alejandroauedevdl.dfs.core.windows.net/'
```

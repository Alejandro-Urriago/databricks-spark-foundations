
### Create managed table in the gold table

#### join driver results to identity the number of wins per driver

```sql
DROP TABLE IF EXISTS f1_dev.gold.driver_wins;

CREATE TABLE IF NOT EXISTS f1_dev.gold.driver_wins
AS
SELECT
  d.name,
  count(1) AS number_of_wins
FROM f1_dev.silver.drivers AS d
INNER JOIN f1_dev.silver.results AS r
  ON d.driver_id = r.driver_id

WHERE r.position = 1
GROUP BY d.name;


```

```sql
SELECT * FROM f1_dev.gold.driver_wins ORDER BY number_of_wins DESC;
```

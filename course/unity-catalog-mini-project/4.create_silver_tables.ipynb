
### Create managed tables in the silver schema

1. drivers
1. results

```sql
DROP TABLE IF EXISTS f1_dev.silver.drivers;


CREATE TABLE IF NOT EXISTS f1_dev.silver.drivers
AS

SELECT
  driverId AS driver_id,
  driverRef AS driver_ref,
  number,
  code,
  concat(name.forename, ' ', name.surname) AS name,
  dob,
  nationality,
  current_timestamp() AS ingestion_time

FROM f1_dev.bronze.drivers;

```

```sql
DROP TABLE IF EXISTS f1_dev.silver.results;


CREATE TABLE IF NOT EXISTS f1_dev.silver.results
AS

SELECT
  resultId AS result_id,
  raceId AS race_id,
  driverId AS driver_id,
  constructorId AS constructor_id,
  number,
  grid,
  position,
  positionText AS position_text,
  positionOrder AS position_order,
  points,
  laps,
  time,
  milliseconds,
  fastestLap AS fastest_lap,
  rank,
  fastestLapTime AS fastest_lap_time,
  fastestLapSpeed AS fastest_lap_speed,
  statusId AS status_id,
  current_timestamp() AS ingestion_time

FROM f1_dev.bronze.results;
```
